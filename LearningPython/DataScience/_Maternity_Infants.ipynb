```python
# function for cleaning the info columns

def sleep_info(sleep):
    return (sleep
            [info]
            .astype({'Participant_number':'int16', 'Type_parents':'int8', 'Birth_1mth_M_inclusion':'int8', 'Birth_12mth_M_inclusion':'int8', 
                    'Age':'int8', 'Marital_status':'int8', 'Marital_status_edit':'int8', 'Education':'int8', 'Gestationnal_age':'int8', 'Type_pregnancy':'int8',
                    'sex_baby1':'int8', 'Age_bb':'int8', 'night_awakening_number_bb1':'int8', 'how_falling_asleep_bb1':'int8'})  
            # .memory_usage()
            # .sum()
            # .dtypes
)
```

```python
# function for cleaning the survey columns

def sleep_survey(sleep):
    return  (sleep
                [survey]
                .select_dtypes('float')  
                .fillna(0)
                .astype('int16')   
                # .describe()
                # .dtypes
            )
```

```python
# creating variable for survey DataFrame

sleep_s = sleep_survey(sleep)
sleep_s
```

```python
# creating variable for info DataFrame

sleep_i = sleep_info(sleep)
sleep_i
```
