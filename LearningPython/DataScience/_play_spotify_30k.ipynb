```python
import numpy as np
import pandas as pd
```

```python
spotify = pd.read_csv('data/spotify_songs.csv')
spotify.dtypes
```

```python
cols = ['track_id', 'track_name', 'track_artist', 'track_popularity', 'track_album_id', 'track_album_name']

(spotify
    [cols]
    .select_dtypes(int)
    .describe()
)
```

```python
spotify = spotify.dropna()
```

```python
spotify.duplicated().sum()
```

```python
spotify.drop(columns=['track_id', 'track_album_id'], inplace=True)
```

```python
spotify.head()
```

```python
spotify.sort_values('track_popularity', inplace=True)
spotify
```

```python
spotify.set_index(['playlist_genre', 'playlist_subgenre'], inplace=True)
```

```python
spotify.sort_values('duration_ms')
```

```python
# playing with strings a little here

spotify[spotify['track_name'].str.contains('ROXANNE')]
spotify.loc[spotify['track_name'].str.contains('roxanne', case=False, regex=True)]      # this is the same as above, note that case=False ignores casing, and regex would be
                                                                                        # correct if we used regex characters
```

```python
spotify[~spotify['track_name'].str.contains('ROXANNE')]     # same as above but uses tilda key to get everyone EXCEPT ROXANNE
```
