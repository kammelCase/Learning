```python
# example data

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
# example data

df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
df
```

```python
# notice how we just get an object as a return
# the groupby function doesn't do the computation until you ask it to

df.groupby('key')
```

```python
df.groupby('key').sum()
```

```python
# more functions
# aggregate(), filter(), transform(), apply()

# example data
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
# aggregate() uses some black magic and let's us use strings, functions, or a list of both...
# some serious magic going on here and i like it

df.groupby('key').aggregate(['min', np.median, max])
```

```python
# can also use aggregate() to pass a dictionary to reorganize data

df.groupby('key').aggregate({'data1':min,
                            'data2':'max'})
```

```python
# filtering
# we can make a function that filters results
# mark: need to do more research on this one, vtd is pretty vague on it

def filter_func(x):
    return x['data2'].std() > 4

df.groupby('key').filter(filter_func)
```

```python
# transform ()
# a function is performed, but the shape of the data is maintained
# useful for centering data as seen below

df.groupby('key').transform(lambda x: x - x.mean())
```

```python
# apply()
# lets you take a DataFrame and apply an arbitrary funciton

def norm_by_data2(x):
    # x is a DataFrame of group values
    x['data1'] /= x['data2'].sum()
    return x

display('df', "df.groupby('key').apply(norm_by_data2)")
```

```python
# note that a key can be any list as long as it's the same length of the index

L = [0, 1, 0, 1, 2, 0]
display('df', 'df.groupby(L).sum()')
```

```python
# we can also use a dictionary as a way to map index values!

df2 = df.set_index('key')
mapping = {'A': 'vowel', 'B':'consonant', 'C':'consonant'}
display('df2', 'df2.groupby(mapping).sum()')
```

```python
# we can actually pass any python function within groupby()
# lowercase strings for key

df2.groupby(str.lower).mean()
```

```python
# we can combine functions as a group on groupby, too

df2.groupby([str.lower, mapping]).mean()
```
